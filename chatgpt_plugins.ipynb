```python
from langchain_openai import ChatOpenAI
from langchain.agents import load_tools, initialize_agent
from langchain.agents import AgentType
from langchain.tools import AIPluginTool
```

Read the ai-plugin.json to get info of how to use the API and what output to expect

```python
from langchain.tools import AIPluginTool

tool = AIPluginTool.from_plugin_url("https://www.wolframalpha.com/.well-known/ai-plugin.json")

# Print all attributes of the tool
print("Tool attributes:")
for key, value in tool.__dict__.items():
    print(f"{key}: {value}")

# Try to access some common attributes
print("\nTrying to access common attributes:")
attributes_to_try = ['name', 'description', 'plugin_spec', 'openapi_spec', 'api_spec']
for attr in attributes_to_try:
    try:
        value = getattr(tool, attr)
        print(f"{attr}: {value}")
    except AttributeError:
        print(f"{attr}: Not available")
```

```python
llm = ChatOpenAI(temperature=0)

# Load the default tools
# tools = load_tools(["requests_all"])

# add the wolfram alpha plugin tool
# tools += [tool]

# initialize the agent
agent_chain = initialize_agent(
    tool, 
    llm, 
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION, 
    verbose=True,
    allow_dangerous_tools=True  # Only add this if absolutely necessary
)
agent_chain.run("what is the population of malaysia?")
```
